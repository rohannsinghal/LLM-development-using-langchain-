```python
# Import necessary modules
from langchain.document_loaders.generic import GenericLoader
from langchain.document_loaders.parsers import OpenAIWhisperParser
from langchain.document_loaders.blob_loaders.youtube_audio import YoutubeAudioLoader
import openai

# Set the OpenAI API key
openai.api_key = 'sk-proj-xxxxxxxxxxxxxxxxxxxx7O'

# Define the YouTube video URL and save directory
url = "https://www.youtube.com/watch?v=jGwO_UgTS7I"
save_dir = "docs/youtube/"

# Create the loader with YouTube audio loader and OpenAI Whisper parser
loader = GenericLoader(
    YoutubeAudioLoader([url], save_dir),
    OpenAIWhisperParser()
)

# Load and transcribe the documents
docs = loader.load()

# Print the first 500 characters of the first document
if docs:
    print(docs[0].page_content[:500])
else:
    print("No documents were loaded.")
```
